```python
import numpy as np
import matplotlib.pyplot as plt
hours_studied = np.array([1, 2, 4, 8])
marks_secured = np.array([15, 70, 60, 95])
a, b = np.polyfit(hours_studied, marks_secured, 1)
print(f"The regression equation is: Y = {a:.2f} + {b:.2f}X")
predicted_marks = np.polyval([a, b], 7)
print(f"A student can score approximately {predicted_marks:.2f} marks if they study for 7 hours.")
plt.scatter(hours_studied, marks_secured, label="Given data")
plt.plot(hours_studied, a + b * hours_studied, color="red", label="Regression line")
plt.xlabel("Hours studied")
plt.ylabel("Marks secured")
plt.legend()
plt.show()
```
